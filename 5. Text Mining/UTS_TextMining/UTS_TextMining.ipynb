##Text Cleaning

```python
import re

def cleansing(df):
    df_clean = df.str.lower() #Mengubah semua karakter menjadi huruf kecil
    df_clean = [re.sub(r"\d+", "", i) for i in df_clean] #Menghapus semua digit
    df_clean = [re.sub(r'[^\w]|com', ' ', i) for i in df_clean] # Menghapus semua kata .com dan non karakter
    df_clean = [re.sub(r'\s+', ' ', i) for i in df_clean] #Menggantikan semua whitespace yang lebih dari satu menjadi single space
    return df_clean
```

```python
clean_text = cleansing(df['Teks'])
```

```python
for i in clean_text:
    print(i)
```

###Tokenization

```python
nltk.download('punkt')
```

```python
word_token=[word_tokenize(i) for i in clean_text]
```

```python
word_token
```

```python
data_enc=[]
for i in word_token:
    data_enc=data_enc+i
```

```python
data_enc
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

Karena teks yang digunakan disini adalah bahasa Indonesia, maka stopwords yang akan digunakan adalah bahasa indonesia.

```python
list_stopwords=set(stopwords.words('indonesian'))
```

```python
list_stopwords
```

```python
token_without_stopwords=[word for word in data_enc if not word in list_stopwords]
```

```python
print(token_without_stopwords)
```

##One Hot Encoding

```python
from numpy import array
from numpy import argmax
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

# integer encode
label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(token_without_stopwords)
print(integer_encoded)
# binary encode
onehot_encoder = OneHotEncoder(sparse=False)
integer_encoded = integer_encoded.reshape(len(integer_encoded), 1)
onehot_encoded = onehot_encoder.fit_transform(integer_encoded)
print(onehot_encoded)
# invert first example
inverted = label_encoder.inverse_transform([argmax(onehot_encoded[0, :])])
print(inverted)
```

Di dalam one hot encoding, kita akan mentranformasi variabel kategori menjadi vektor biner. 

###CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
stop_words = stopwords.words('indonesian')
```

```python
vectorizer = CountVectorizer(max_features=50, stop_words=stop_words, ngram_range=(1, 1), min_df=2)
X = vectorizer.fit_transform(clean_text)
CVec=pd.DataFrame(X.toarray(),columns=vectorizer.get_feature_names_out())
CVec.head()
```

Pada proses ini, kita akan mengubah teks pada data menjadi vektor dengan mengitung frekuensi kemunculan kata pada data. Karena size vektor yang diinginkan adalah 50 dan kata yang diperhitungkan minimal 2, maka akan kita gunakan 'max_features=50' dan 'min_df=2'. 

Kemudian kita gunakan data yang sudah diberishkan yaitu "clean_text" untuk feature extraction.

Dari hasil di atas, kita bisa melihat frekuensi kemunculan kata-kata yang ada di dalam data kita.

##TF-IDF

Pada bagian ini, kita akan mencari bobot dari kata-kata yang paling sering muncul pada data kita yang memiliki. TF-IDF dapat menunjukkan jika kata tersebut penting atau tidak dalam konteks data kita. 



```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(max_features=50, min_df=2, stop_words=stop_words)
X = vectorizer.fit_transform(clean_text)
#vectorizer.get_feature_names_out()
TFIDF=pd.DataFrame(X.toarray(),columns=vectorizer.get_feature_names_out())
TFIDF.head()
```

Kita akan menggunakan tfidfVectorizer untuk mencari nilai TF-IDF dari data kita. Disini kita akan menggunakan max_features=50 dan min_df=2 agar hasilnya konsisten dengan proses sebelumnya. 

```python
for i, row in TFIDF.iterrows():
    top_tfidf = row.nlargest(5)
    print(f"\nData ke-{i+1}:")
    for word, score in top_tfidf.items():
        print(f"{word}: {score:.2f}")
```

Berikut adalah 5 nilai TF-IDF tertinggi dari masing-masing teks di dalam data.
